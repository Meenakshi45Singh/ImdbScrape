```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
source=requests.get("https://www.imdb.com/chart/top/")
source.raise_for_status
    
soup=BeautifulSoup(source.text,"html.parser")
movies=soup.find("tbody",{"class":"lister-list"}).findAll('tr')
```

```python
name=[]
director=[]
release=[]
stars=[]

for movie in movies:
        name.append(movie.find("td",{"class":"titleColumn"}).a.text)
        release.append(movie.find("td",{"class":"titleColumn"}).span.text.strip('()'))
        director.append(movie.find("td",{"class":"titleColumn"}).a['title'].split(",")[0].replace(" (dir.)",""))
        stars.append(movie.find("td",{"class":"titleColumn"}).a['title'].split(",")[1]+','+movie.find("td",{"class":"titleColumn"}).a['title'].split(",")[2])
        
```

```python
movie_DF=pd.DataFrame({'Name':name,'Release':release,'Director':director,'Stars':stars})
```
